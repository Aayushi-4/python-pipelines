Remove prefixes or suffixes using nlp spacy

```python
import spacy
```

```python
nlp=spacy.load("en_core_web_sm")
```

```python
def clean_name(name):
    doc=nlp(str(name))
    letters=[token.text for token in doc
             if token.ent_type_=='PERSON' or token.pos_=='PROPN']
    return' '.join(letters)
```

```python
df['clean_name']=df['name'].apply(clean_name)
```

```python
df['first_name']=df['clean_name'].str.split().str[0]
df['last_name']=df['clean_name'].str.split().str[-1]
```
