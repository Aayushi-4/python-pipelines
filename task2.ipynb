```python
import pandas as pd
```

```python
df_customer=pd.read_csv(r"C:\Users\Aayushi\Downloads\us_customer_data 1.csv")
df_customer.head()
```

```python
df_customer.isnull().sum()
```

```python
df_customer['email']=df_customer['email'].fillna(df_customer['name'].str.replace(' ','').str.lower()+'@unkown.com')
```

```python
df_customer.isnull().sum()
```

```python
df_customer.columns=df_customer.columns.str.strip()
```

```python
df_customer['name'] = df_customer['name'].str.strip().str.title()
```

```python
df_customer
```

```python
# Remove non-numeric characters
df_customer['phone'] = df_customer['phone'].astype(str).str.replace(r'\D', '', regex=True)

# Optional: Fill missing or short numbers
df_customer['phone'] = df_customer['phone'].apply(lambda x: x if len(x) == 10 else pd.NA)


```

```python
df_customer['phone']=df_customer['phone'].fillna('0123456789')
```
